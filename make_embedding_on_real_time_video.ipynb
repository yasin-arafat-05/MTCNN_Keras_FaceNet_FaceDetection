```python
import cv2 as cv 
from mtcnn import MTCNN
from keras_facenet import FaceNet
```

```python
detector = MTCNN()
embedder = FaceNet()

def verify_faces(image,boxes):
    faces = []
    img_width, img_height = image.shape[:2]
    for box in boxes:
        x,y,width,height = box 
        # ensuring bounding box in your image frame: 
        x = min(x,img_width)
        y = min(y,img_height)
        # append the face in the faces[] list
        faces.append(image[y:y+height, x:x+width])
    return faces
    
    
    
cap = cv.VideoCapture(0)
while cap.isOpened():
    embeddings = []
    ret ,frame = cap.read()
    # Convert frame to RGB as MTCNN expects RGB images
    rgb_frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    output_frame = detector.detect_faces(rgb_frame)
    if output_frame:
        boxes = [face['box'] for face in output_frame]
        faces = verify_faces(rgb_frame,boxes)
        embeddings.append(embedder.embeddings(faces))
        
    print(f"Generated {len(embeddings)} embeddings.")

    for i, embedding in enumerate(embeddings):
        print("Current_Embedding: {}".format(embedding))
        x1, y1, width, height = boxes[i]
        x2, y2 = x1 + width, y1 + height
        # Draw rectangle and label on the original frame
        cv.rectangle(rgb_frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv.putText(rgb_frame, "Yasin Arafat", (x1, y1 - 10),
            cv.FONT_HERSHEY_SIMPLEX,
            0.9, (0, 255, 0), 2)

    cv.imshow("output",rgb_frame)
    
    if cv.waitKey(1) & 0xff == ord('q'):
        break
cap.release()
cv.destroyAllWindows()



```
